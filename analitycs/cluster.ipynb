```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('Участники anonimized.csv', sep=',')
data
```

```python
import datetime

# преобразуем строковые значения в тип datetime
data['Начало трудового стажа'] = pd.to_datetime(data['Начало трудового стажа'], errors='coerce')

# вычисляем количество лет до сегодняшней даты
now = datetime.datetime.now().date()
data['Трудовой стаж,лет'] = (now - data['Начало трудового стажа']).astype('<m8[Y]')
data
```

```python
data.fillna(0)
```

```python
import datetime

now = datetime.datetime.now().date()
data['Дата рождения'] = pd.to_datetime(data['Дата рождения'], errors='coerce')
data['Дата рождения']=data['Дата рождения'].dt.date
data['Возраст'] = (now - data['Дата рождения']).astype('timedelta64[Y]')
data['Начало трудового стажа']=pd.to_datetime(data['Начало трудового стажа'], errors='coerce').dt.date
data['Стаж'] = (now - data['Начало трудового стажа']).astype('timedelta64[Y]')
data['Начало трудовой деятельности в РОСАТОМ']=pd.to_datetime(data['Начало трудовой деятельности в РОСАТОМ'], errors='coerce').dt.date
data['Стаж в РосАтом'] = (now - data['Начало трудовой деятельности в РОСАТОМ']).astype('timedelta64[Y]')
data=data.fillna(0)
```

```python
features_cluster=['Стаж в РосАтом','Возраст','Стаж']
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1.6, min_samples=3).fit(data[features_cluster])
labels = dbscan.labels_
data['cluster'] = dbscan.labels_
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
def plot_3d(df, labels, title):
    reduced_data = np.array(df)
    
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection=Axes3D.name)
    
    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k','kg']
    
    for i, label in enumerate(np.unique(labels)):
        ax.scatter(reduced_data[labels == label, 0], reduced_data[labels == label, 1],
                   reduced_data[labels == label, 2], cmap='viridis', label=label)
    
    ax.legend()
    ax.set_title(title)

plot_3d(data[features_cluster][data['cluster']!=-1], data['cluster'][data['cluster']!=-1], title='Кластеризация')
plt.show()
```
