```python
import seaborn as sns
import matplotlib.pyplot as plt
def grah(compet,place_work):
    df=pd.read_csv(f'/home/jupyter/RosAtom/222/{compet}.csv',sep=',')
    df['Баллы, %']=df['Баллы, %'].str.replace(',','.')
    ax=sns.barplot(x=df['ФИО'],y=df['Баллы, %'].astype('float'))
    plt.xlabel(f'Баллы участников {place_work}')
    ax.figure.savefig("/home/jupyter/RosAtom/111.jpeg", dpi=300)
```
